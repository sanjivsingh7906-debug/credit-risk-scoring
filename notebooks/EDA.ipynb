```python
df_clean = df.copy()
print(df_clean.duplicated().sum())
```

```python
df_clean.duplicated().sum()
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.shape
```

```python
import numpy as np

df_clean["months_on_file"] = df_clean["months_on_file"].replace(-1, np.nan)
```

```python
df_clean["months_on_file"].isna().sum()
```

```python
print(df_clean[df_clean["months_on_file"]==-1])
```

```python
df_clean["credit_utilization"].map(type).value_counts()
```

```python
df_clean["credit_utilization"] = (
    df_clean["credit_utilization"]
    .astype(str)
    .str.replace("%", "", regex=False)
)

df_clean["credit_utilization"] = pd.to_numeric(
    df_clean["credit_utilization"],
    errors="coerce"
)
```

```python
df_clean.loc[df_clean["credit_utilization"] > 1, "credit_utilization"] /= 100
```

```python
df_clean["credit_utilization"].describe()
```

```python
df_clean["income"] = (
    df_clean["income"]
    .astype(str)
    .str.lower()
    .str.replace(",", "", regex=False)
)
```

```python
df_clean["income"] = pd.to_numeric(
    df_clean["income"],
    errors="coerce"
)
```

```python
df_clean["income"].describe()
```

- Age shows a non-monotonic, non-linear relationship with default risk, with elevated risk in middle-to-older age ranges. Signal is weak and unstable due to small sample size.
- Income shows a monotonic but counter-intuitive relationship with default in this sample, likely confounded by loan size or loan type. Signal should be interpreted cautiously.
- While individual exceptions exist, income shows a consistent monotonic trend across quantiles, indicating a robust directional signal.
